### **1. What is the difference between interpreted and compiled languages?**



Interpreted languages run the program line by line, while compiled languages convert the entire code into machine language before running. Interpreted languages are easier to test and debug, while compiled languages run faster after compilation.

### **2. What is exception handling in Python?**

Exception handling is the process of dealing with errors during program execution so the program doesn’t crash. It allows the program to handle unexpected situations smoothly.

<br>

---
<br>

### **3. What is the purpose of the `finally` block in exception handling?**

The `finally` block always runs after the `try` block, whether an error occurs or not. It is usually used to clean up resources like closing files or disconnecting from databases.

<br>

<br>

### **4. What is logging in Python?**

Logging means recording events that happen while the program runs. It helps in debugging and tracking the flow of the program without printing everything manually.

<br>

### **5. What is the significance of the `__del__` method in Python?**

The `__del__` method is called when an object is deleted. It can be used to clean up resources like closing files or releasing memory before the object is removed.

### **6. What is the difference between `import` and `from ... import` in Python?**

The `import` keyword brings in the whole module, while `from ... import` brings in only specific functions or variables. This makes the code cleaner and more focused when only part of a module is needed.

Multiple exceptions can be handled by using separate `except` blocks for different error types, or by combining them in one block. This helps manage specific errors properly.

### **8. What is the purpose of the `with` statement when handling files in Python?**

The `with` statement ensures that the file is automatically closed after the block of code is finished, even if an error occurs. It makes file handling safer and cleaner.

### **9. What is the difference between multithreading and multiprocessing?**

Multithreading allows multiple tasks to run at the same time within the same program, sharing memory. Multiprocessing runs tasks in separate processes, each with its own memory. Multiprocessing is better for heavy CPU tasks.

Logging helps in debugging, tracking, and understanding the behavior of the program. It allows developers to store information about warnings, errors, and important events in files.

Memory management in Python means how memory is allocated and released. Python handles most of it automatically using garbage collection, so developers don’t need to free memory manually.

Python keeps track of how many times an object is being used. When an object is no longer needed, Python automatically deletes it to free memory. This process is called garbage collection.
